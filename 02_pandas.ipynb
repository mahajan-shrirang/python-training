```python
def convert_pclass(x):
    if x == 1:
        return 'First Class'
    elif x == 2:
        return 'Second Class'
    else:
        return 'Third Class'
```

```python
df['Pclass'] = df['Pclass'].apply(convert_pclass)
```

```python
df
```

```python
pclasses = {
    'First Class': 1,
    'Second Class': 2,
    'Third Class': 3
}
```

```python
df['Pclass'] = df['Pclass'].map(pclasses)
```

```python
# Adds two numbers
def add(a: int | float, b: int | float) -> int | float:
	"""Adds two numbers

	Args:
		a (int | float): The first number
		b (int | float): The second number

	Returns:
		int | float: The sum of the two numbers
	"""
	return a + b
```
